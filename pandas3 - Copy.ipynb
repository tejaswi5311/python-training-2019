```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
df = pd.DataFrame(np.random.randint(45,100,size=(2000,1)),columns=list('m'))
df
fc=0
sc=0
tc=0
fc=0
cl=[]
for i in df.m:
    if i>=80:
        fc+=1
    elif i>=75 and i<=80:
        sc+=1
    elif i>=65 and i<=75:
        tc+=1
    else:
        fc+=1
cl.append(fc)
cl.append(sc)
cl.append(tc)
cl.append(fc)
a=['fc','sc','tc','fc']
print(a,cl)
plt.pie(cl,labels=a,autopct='%4.1f%%')
plt.legend(loc="upper left")
plt.show()
        

```
