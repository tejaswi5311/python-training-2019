```python
a=0
b=0
c=0
d=0
w=[]
for i in f.cont_rating:
    if i=="4+":
        a+=1
    elif i=="9+":
        b+=1
    elif i=="12+":
        c+=1
    elif i=="17+":
        d+=1
    else:
        pass
w.append(a)
w.append(b)
w.append(c)
w.append(d)
print("num of apps are 4+:",a)
print("num of apps are 9+:",b)
print("num of apps are 12+:",c)
print("num of apps are 17+:",d)
            
        
```

```python
import matplotlib.pyplot as plt
l=[]
for i in f.cont_rating:
    if i not in l:
        l.append(i)
print(l,w)
plt.pie(w,labels=l,autopct='%4.1f%%')
plt.legend(loc="upper left")
        

```
